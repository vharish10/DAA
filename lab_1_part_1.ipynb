```python
#%%
import time
import matplotlib.pyplot as plt

def recursive_sumofN(n):
    if n == 0:
        return 0
    else:
        return n + recursive_sumofN(n-1)
    
def iterative_sumofN(n):
    Sum = 0
    for i in range(n+1):
        Sum +=i
    return Sum

def time_comparison(func, n):
    start =time.time()
    func(n)
    end = time.time()
    return end-start

n_values=list(range(1,100))

recursive_times = [time_comparison(recursive_sumofN, n) for n in n_values]
iterative_times = [time_comparison(iterative_sumofN, n) for n in n_values]

plt.plot(n_values, recursive_times, label='recursive')
plt.plot(n_values, iterative_times, label='iterative')
plt.xlabel('N')
plt.ylabel('time')
plt.legend()
plt.show()




#%%
import random


# Define the linear search function
def linear_search(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1

# Define the binary search function
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    mid = 0

    while low <= high:
        mid = (high + low) // 2
        if arr[mid] < x:
            low = mid + 1
        elif arr[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1

# Generate an array of 10000 elements in the range 1 to 1000
arr = [random.randint(1, 1000) for _ in range(10000)]
arr.sort()

# Define the search keys
search_keys = [random.randint(1, 1000) for _ in range(5)]

# Initialize lists to store the time taken by each algorithm
linear_search_times = []
binary_search_times = []

# Perform the searches and measure the time taken
for key in search_keys:
    start_time = time.time()
    linear_search(arr, key)
    end_time = time.time()
    linear_search_times.append(end_time - start_time)

    start_time = time.time()
    binary_search(arr, key)
    end_time = time.time()
    binary_search_times.append(end_time - start_time)

# Plot the results
plt.plot(search_keys, linear_search_times, label='Linear Search')
plt.plot(search_keys, binary_search_times, label='Binary Search')
plt.xlabel('Search Key')
plt.ylabel('Time Taken (seconds)')
plt.legend()
plt.show()

#%%
def string_to_int(s):
    if len(s) == 1:
        return s
    else:
        return string_to_int(s[:-1]) + s[-1]

# Ask the user to enter a string of digits
s = input("Please enter a string of digits: ")

# Convert the string to an integer
integer = string_to_int(s)

# Format the integer with comma separation
formatted_integer = "{:,}".format(int(integer))

# Print the result
print(f"The string '{s}' represents the integer {formatted_integer}")

#%%
def reverse_string(s):
    if len(s) == 0:
        return s
    else:
        return reverse_string(s[1:]) + s[0]

# Ask the user to enter a string
s = input("Please enter a string: ")

# Print the reverse of the string
print(f"The reverse of '{s}' is '{reverse_string(s)}'")

#%%
def is_palindrome(s):
    # Base case: if the string is empty or has only one character, it is a palindrome
    if len(s) < 2:
        return True
    # If the first and last characters are not the same, the string is not a palindrome
    if s[0] != s[-1]:
        return False
    # Recursive case: check if the middle part of the string is a palindrome
    return is_palindrome(s[1:-1])

# Ask the user to enter a string
s = input("Please enter a string: ")

# Check if the string is a palindrome and print the result
if is_palindrome(s):
    print(f"'{s}' is a palindrome.")
else:
    print(f"'{s}' is not a palindrome.")
```

```python
import random
import time
import matplotlib.pyplot as plt

# Define the sorting algorithms
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Generate 1000 random integers between 1 and 10000
arr = [random.randint(1, 10000) for _ in range(1000)]

# Initialize lists to store the time taken by each algorithm
bubble_sort_times = []
selection_sort_times = []
insertion_sort_times = []

# Perform the sorting and measure the time taken
start_time = time.time()
bubble_sort(arr.copy())
end_time = time.time()
bubble_sort_times.append(end_time - start_time)

start_time = time.time()
selection_sort(arr.copy())
end_time = time.time()
selection_sort_times.append(end_time - start_time)

start_time = time.time()
insertion_sort(arr.copy())
end_time = time.time()
insertion_sort_times.append(end_time - start_time)

# Plot the results
plt.plot(['Bubble Sort', 'Selection Sort', 'Insertion Sort'], [bubble_sort_times[0], selection_sort_times[0], insertion_sort_times[0]])
plt.ylabel('Time Taken (seconds)')
plt.show()
```
